```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

# requested page source from

```python
url="https://www.flipkart.com/search?q=mobile%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
page=requests.get(url)
page
```

```python
#display the page source code
page.content
```

```python
soup= bs(page.content,"html.parser")
print(soup.prettify())
```

# Title of website

```python
soup.title
```

```python
Title= soup.title
print(Title.string)
```

```python
box= soup.find('div',class_='_1YokD2 _3Mn1Gg')
```

```python
box
```

# product name

```python
name=box.find_all('div',class_='_4rR01T')
name
```

```python
product_name= []

for i in range(0,len(name)):
    product_name.append(name[i].get_text())
```

```python
product_name
```

```python
print(len(name))
```

# product price

```python
price=box.find_all('div',class_='_30jeq3 _1_WHN1')
price
```

```python
product_price= []

for i in range(0,len(price)):
    product_price.append(price[i].get_text())
```

```python
product_price
```

```python
print(len(price))
```

# product features

```python
features=box.find_all('ul',class_='_1xgFaf')
features
```

```python
product_features= []

for i in range(0,len(features)):
    product_features.append(features[i].get_text())
```

```python
product_features
```

```python
print(len(product_features))
```

# DataFrame

```python
d = {"product_name":product_name,
   "price":product_price,
   "features":product_features,
   "rating":product_rating}
```
